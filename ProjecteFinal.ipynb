**2:** Gràfiques

**2.1:** *1a gràfica*:  Vendes de Cotxes per model 

* Tornem a crear una variable a la que li donarem l'equivalent a les columnes que utilitzarem

NOTA: Aquest procés de comentar la mateixa comanda no l'estaré repetint cada 2x3 així que cada vegada que es vegi algo com la següent comanda significarant totes el mateix

```python
ventas_por_modelo = cars_dataset.groupby('model')['price'].sum().reset_index()
```

* Ara, printem la propia gràfica

```python
plt.figure(figsize=(10, 6))
plt.bar(ventas_por_modelo['model'], ventas_por_modelo['price'])
plt.xlabel('Marca de coche')
plt.ylabel('Ventas')
plt.title('Ventas de coches por modelo')
```
